```python
%load_ext autoreload
%autoreload 2

from brancharchitect.jumping_taxa.lattice.lattice_construction import construct_sub_lattices
from brancharchitect.jumping_taxa.tree_interpolation import interpolate_tree
from brancharchitect.io import parse_newick

t1, t2 = parse_newick(("((A,B),(C,(E,F)));" + "((((A,B),E),C),F);"))
print(t1.to_newick())
print(t2.to_newick())
s_edge_container = construct_sub_lattices(t1, t2)
for edge in s_edge_container:
    print("Arm One")
    for component in edge.left_cover:
        print(component)
    print("Arm Two")
    for component in edge.right_cover:
        print(component)
```

```python
%load_ext autoreload
%autoreload 2

from brancharchitect.partition_set import Partition, PartitionSet

# Suppose the order of taxa is fixed.
order = ("A", "B", "C", "D")
# Create a few IndexedSplit objects.
split1 = Partition((0, 2), order)
split1 = Partition((0, 1), order)
split2 = Partition((2, 3), order)
split3 = Partition((0, 2), order)
split4 = Partition((1, 3), order)
# Create two IndexedSplitSet instances.
set1 = PartitionSet({split1, split2, split3}, order=order, name="Set1")
set2 = PartitionSet({split2, split3, split4}, order=order, name="Set2")
# Now you can use set operations and extra methods:
print("Set1:", set1)
print("Set2:", set2)
```
