```python
from Bio import Entrez, SeqIO


def fetch_genbank_sequences(accession_numbers):
    """
    Fetch sequences from GenBank using a list of accession numbers.

    Parameters:
    - accession_numbers: List of GenBank accession numbers to fetch.

    Returns:
    - sequences: List of SeqRecord objects representing the fetched sequences.
    """
    Entrez.email = "your.email@example.com"  # Replace with your email
    sequences = []

    # Fetch sequences from GenBank using accession numbers
    for acc in accession_numbers:
        handle = Entrez.efetch(db="nucleotide", id=acc, rettype="gb", retmode="text")
        seq_record = SeqIO.read(handle, "genbank")
        sequences.append(seq_record)
        handle.close()

    return sequences


# List of accession numbers for Norovirus sequences
accession_numbers = [
    "KR074148",
    "KR074149",
    "KR074150",
    "KR074151",
    "KR074152",
    "KR074153",
    "KR074154",
    "KR074155",
    "KR074156",
    "KR074157",
    "KR074158",
    "KR074159",
    "KR074160",
    "KR074161",
    "KR074162",
    "KR074163",
    "KR074164",
    "KR074165",
    "KR074166",
    "KR074167",
    "KR074168",
    "KR074169",
    "KR074170",
    "KR074171",
    "KR074172",
    "KR074173",
    "KR074174",
    "KR074175",
    "KR074176",
    "KR074177",
    "KR074178",
    "KR074179",
    "KR074180",
    "KR074181",
    "KR074182",
    "KR074183",
    "KR074184",
    "KR074185",
    "KR074186",
    "KR074187",
    "KR074188",
    "KR074189",
    "KR074190",
    "KR074191",
]

# Fetch the sequences
norovirus_sequences = fetch_genbank_sequences(accession_numbers)

# Save sequences to a FASTA file
with open("norovirus_sequences.fasta", "w") as output_handle:
    SeqIO.write(norovirus_sequences, output_handle, "fasta")

print("Sequences have been saved to 'norovirus_sequences.fasta'")
```
